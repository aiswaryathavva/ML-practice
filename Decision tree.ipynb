```python
#decision Tree:Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly it is preferred for solving Classification problems. It is a tree-structured classifier, where internal nodes represent the features of a dataset,
#  branches represent the decision rules and each leaf node represents the outcome.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
iris.columns
```

```python
iris.plot(kind="density",subplots=True,layout=(3,3))
```

```python
import seaborn as sns
plt.figure(figsize=(15,9))
sns.heatmap(iris.corr(),cmap="YlGnBu")
plt.show()
# ,anot=True
# import seaborn as sns
# fig = plt.figure(figsize=(12,12))
# r = sns.heatmap(flight_matrix, cmap='BuPu')
```

```python
x=iris.iloc[:,1:5].values
y=iris["Species"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion="gini")
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn import tree
plt.figure(figsize=(30,25))
dot_data=tree.plot_tree(classifier,feature_names=["Id",'SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
                        class_names=['Iris-setosa', 'Iris-versicolor','Iris-virginica'],filled=True)
plt.title("Decision Tree Visualisation")
plt.show()
```
